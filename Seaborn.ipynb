**Matplotlib**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,30]

plt.plot(x,y)
plt.show()
```

**Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(X=x,Y=y)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.scatterplot(data=tips,x="total_bill",y="tip")
plt.title("Total bill vs Tip")
plt.grid(True)
plt.show()
```

 **Histplot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [45,50,52,55,58,60,61,65,70,72,75,80,85,90]

sns.histplot(marks)

plt.show()
```

**Parameters**
**bins**

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks=[45,50,52,55,58,60,61,65,70,72,75,80,85,90]

sns.histplot(
marks,
bins=6,
color="skyblue",
edgecolor="black",
kde=True
)

plt.title("Marks Distribution")

plt.show()
```

**KDE**

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks=[45,50,52,55,58,60,61,65,70,72,75,80,85,90]

sns.kdeplot(marks)

plt.show()
```

**Parameters**
**fill**=**True**

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks=[45,50,52,55,58,60,61,65,70,72,75,80,85,90]

sns.kdeplot(
marks,
fill=True,
color="green",
linewidth=2
)

plt.title("Marks Density")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({

"Department":["IT","IT","IT","HR","HR","Sales","Sales"],

"Salary":[50000,55000,60000,30000,35000,45000,47000],

"Gender":["Male","Female","Male","Male","Female","Male","Female"]

})

sns.swarmplot(

data=df,

x="Department",

y="Salary",

hue="Gender",

palette="Set2",

size=7

)

plt.title("Department Salary Distribution")

plt.grid(True)

plt.show()
```

**Heatmap**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = [

[10,20,30],

[40,50,60],

[70,80,90]

]

df = pd.DataFrame(data)

sns.heatmap(df)

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({

"Age":[21,22,23,24,25],

"Experience":[1,2,3,4,5],

"Salary":[25000,30000,40000,50000,60000]

})

corr = df.corr()

sns.heatmap(

corr,

annot=True,

cmap="cool",

linewidths=1,

fmt=".2f"

)

plt.title("Correlation Heatmap")

plt.show()
```

**Seaborn** **Themes** & **Styling**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()

tips = sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_theme(

style="whitegrid",

palette="Set2",

context="talk"

)

plt.figure(figsize=(10,6))

sns.barplot(

data=tips,

x="day",

y="total_bill",

hue="sex"

)

plt.title("Average Bill by Day")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(

style="darkgrid",

palette="deep",

context="notebook"

)

plt.figure(figsize=(12,6))

df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [100, 120, 150, 130, 180]
})

sns.lineplot(

data=df,

x="Month",

y="Sales",

marker="o"

)

plt.title("Monthly Sales")

plt.show()
```

**FacetGrid**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

g = sns.FacetGrid(

tips,

col="time"

)

g.map(

sns.histplot,

"total_bill"

)

plt.show()
```
